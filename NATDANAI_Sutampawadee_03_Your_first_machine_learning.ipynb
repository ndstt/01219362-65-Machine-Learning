```python
from pandas import read_csv
url = "https://archive.ics.uci.edu/static/public/17/data.csv"
data = read_csv(url)
data = data.drop(columns = ['ID'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Assuming 'data' is your DataFrame and 'Diagnosis' is the target variable
X = data.drop('Diagnosis', axis=1)  # Features
y = data['Diagnosis']  # Target variable

X_scaled = scaler.fit_transform(X)

scaled_data = pandas.DataFrame(X_scaled, columns=X.columns, index=X.index)
scaled_data['Diagnosis'] = y  # Add back the target variable
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(30, 8))
sns.boxplot(data=scaled_data)
plt.show()
```

```python
scaled_data.head(10)
```

```python
scaled_data['Diagnosis'] = scaled_data['Diagnosis'].map({'B': 0, 'M': 1})
```

```python
# Calculate the correlation matrix
correlation_matrix = scaled_data.corr()

# Select correlations with 'Diagnosis'
diagnosis_corr = correlation_matrix['Diagnosis']

# Remove 'Diagnosis' from index to avoid self-correlation
diagnosis_corr = diagnosis_corr.drop('Diagnosis')

# Sort the correlations in descending order
diagnosis_corr = diagnosis_corr.sort_values(ascending=False)

# Create the heatmap
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.heatmap(diagnosis_corr.to_frame(), annot=True, cmap='coolwarm', fmt=".2f", cbar=True)

plt.title("Correlation with Diagnosis (Sorted)")
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = scaled_data.corr()

# Select correlations with 'Diagnosis'
diagnosis_corr = correlation_matrix['Diagnosis']

# Remove 'Diagnosis' from index to avoid self-correlation
diagnosis_corr = diagnosis_corr.drop('Diagnosis')

# Filter features with correlation >= 0.5
high_corr = diagnosis_corr[diagnosis_corr.abs() >= 0.5]
high_corr_feature = diagnosis_corr[diagnosis_corr.abs() >= 0.5].index

# Sort the correlations in descending order
new_diagnosis_corr = high_corr.sort_values(ascending=False)
```

```python
# Drop features with correlation < 0.6
X_filtered = scaled_data[high_corr_feature]

# Create the heatmap
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.heatmap(new_diagnosis_corr.to_frame(), annot=True, cmap='coolwarm', fmt=".2f", cbar=True)

plt.title("Correlation with Diagnosis (Sorted)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'scaled_data' is your DataFrame
X = scaled_data.drop('Diagnosis', axis=1)  # Features
y = scaled_data['Diagnosis']  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

models = []
models.append(('RF', RandomForestClassifier()))
models.append(('SVM', SVC(gamma='auto')))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('NN', MLPClassifier(hidden_layer_sizes=4)))
models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
models.append(('NB', GaussianNB()))
```

```python
from sklearn.metrics import accuracy_score
# Initialize variables to track the best model
best_model_name = None
best_model_score = 0
best_model = None

# Evaluate each model
for name, model in models:
    # Fit the model on the training data
    model.fit(X_train, y_train)

    # Predict on the test data
    y_pred = model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)

    # Print model performance
    print(f'{name}: Accuracy = {accuracy:.3f}')

    # Update the best model if current model performs better
    if accuracy > best_model_score:
        best_model_score = accuracy
        best_model_name = name
        best_model = model

# Print the best model's name and accuracy
print(f'\nBest Model: {best_model_name} with Accuracy = {best_model_score:.3f}')
```

```python
print("Selected Model : %s"%(best_model_name))
model = best_model
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)

print(predictions)
```

```python
from sklearn.metrics import accuracy_score
print("accuracy = %s"%(accuracy_score(y_test, predictions)))
```

```python
from sklearn.metrics import confusion_matrix
cfm=confusion_matrix(y_test, predictions)
print(cfm)
sns.heatmap(cfm,annot=True)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
